```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[2, 0.5, 0.5], [0, 5, 1], [1, -1, 8]], dtype = float)
n = A.shape[0]
eigValues = np.linalg.eigvals(A)
print(eigValues)

R = np.zeros(n) # disk radii or the Abs sum of non-diagonal entries
for i in range(n):
    R[i] = np.sum(np.abs(A[i,:])) - np.abs(A[i,i])

fig, ax = plt.subplots()
ax.axis([-1, 12, -4, 4])

for k in range(n):
    x, y = A[k,k].real, A[k,k].imag
    ax.add_artist( plt.Circle((x, y), R[k], alpha=0.3) )
    plt.plot(eigValues[k].real, eigValues[k].imag, 'kp')

ax.set_aspect(1)    
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.title("Gershgorin discs")
```
